```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
paths={
    "circuits":"circuits.csv",
    "constructors":"constructors.csv",
    "constructor_results":"constructor_results.csv",
    "constructors":"constructors.csv",
    "driver_standings":"driver_standings.csv",
    "drivers":"drivers.csv",
    "lap_times":"lap_times.csv",
    "pit_stops":"pit_stops.csv",
    "qualifying":"qualifying.csv",
    "races":"races.csv",
    "results":"results.csv",
    "seasons":"seasons.csv",
    "sprint_results":"sprint_results.csv",
    "status":"status.csv"
}
```

```python
df={
    name:pd.read_csv(path) for name,path in paths.items()
}
df
```

```python
master_df=df["results"].merge(
    df['drivers'],
    on="driverId",
    how="left"
)
```

```python
master_df=master_df.merge(
    df['constructors'],
    on="constructorId",
    how="left"
)
```

```python
master_df=master_df.merge(
    df['races'],
    on="raceId",
    how="left"
)
```

```python
df['circuits'].columns.tolist()
```

```python
circuits_df=df['circuits'].drop(columns=['url'],errors='ignore')
```

```python
master_df=master_df.merge(
   circuits_df,
   on="circuitId",
   how="left"
)
```

```python
master_df=master_df.loc[:,~master_df.columns.duplicated()]
```

```python
if "time_y" in master_df.columns:
    master_df['race_datetime']=pd.to_datetime(
        master_df['date'].astype(str)+' '+master_df['time_y'].fillna(''),
        errors='coerce'
    )
```

```python
drop_cols=[
    'fp1_date','fp1_time','fp2_date','fp2_time',
    'fp3_date','fp3_time','quali_date','quali_time',
    'sprint_date','sprint_time','driver_url','constructor_url'
]
master_df=master_df.drop(columns=[c for c in drop_cols if c in master_df.columns])
```

```python
master_df=master_df.sort_values(['year','round','positionOrder']).reset_index(drop=True)
```
